```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

#Answer 1:

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2)
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

#Answer 2:

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

output_notebook()
np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100

sizes = np.random.rand(100) * 50
colors = np.random.rand(100) * 100

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y')
p.scatter('x', 'y', size='sizes', color=color_mapper, fill_alpha=0.6, source=source)
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]


#Answer 3:

from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

output_notebook()
p = figure(x_range=fruits, height=400, title="Fruit Counts", toolbar_location=None, tools="")
p.vbar(x=fruits, top=counts, width=0.9)
p.xgrid.grid_line_color = None
p.y_range.start = 0
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)

#Answer 4:

import numpy as np
from bokeh.plotting import figure, show

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p_hist = figure(title="Histogram of Data", background_fill_color="#fafafa")
p_hist.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)
p_hist.y_range.start = 0
p_hist.xaxis.axis_label = 'Value'
p_hist.yaxis.axis_label = 'Frequency'
p_hist.grid.grid_line_color = "white"
show(p_hist)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
# data_heatmap = np.random.rand(10, 10)
# x = np.linspace(0, 1, 10)
# y = np.linspace(0, 1, 10)
# xx, yy = np.meshgrid(x, y)


#Answer 5:

from bokeh.palettes import brewer
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.plotting import figure, show

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p_heatmap = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1), tools="")

color_mapper = LinearColorMapper(palette=brewer['YlGnBu'][9], low=data_heatmap.min(), high=data_heatmap.max())
p_heatmap.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p_heatmap.add_layout(color_bar, 'right')
show(p_heatmap)
```
